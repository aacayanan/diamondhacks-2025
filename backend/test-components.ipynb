```python
def get_bpm(r_should):
    time_interval = []
    peak_index = ss.find_peaks(r_should, prominence=0.013)  # 118 bpm
    print(peak_index[0])
    for i in range(len(peak_index[0]) - 1):
        second = (peak_index[0][i + 1] - peak_index[0][i]) / 30
        time_interval.append(second)
    print(time_interval)
    seconds_avg = np.mean(time_interval)
    return (1 / seconds_avg) * 60


get_bpm(r_shoulder_data)
```
